```python
# functions for statistics
def average(arr):
    count = 0
    for i in range(len(arr)):
        count += arr[i]
    return count/len(arr)

def average_n(arr, n):
    count = 0
    for i in range(len(arr)):
        if i != n:
            count += arr[i]
    return count/(len(arr) - 1)

def jackknife_err(arr):
    arr_new = []
    err = 0
    aver = average(arr)
    for i in range(len(arr)):
        arr_new.append(average_n(arr, i))
    for i in range(len(arr_new)):
        err += (aver - arr_new[i])*(aver - arr_new[i])
    err = math.sqrt(err * (len(arr_new) - 1) / len(arr_new))
    return aver, err

def jackknife_electric(correlator, wilson, plaket):
    field_new = []
    field_aver = average(correlator) / average(wilson) - average(plaket)
    
    for i in range(len(correlator)):
        field_new.append(average_n(correlator, i) / average_n(wilson, i) - average_n(plaket, i))
    err = 0
    for i in range(len(field_new)):
        err += (field_aver - field_new[i]) * (field_aver - field_new[i])
    err = math.sqrt(err * (len(field_new) - 1) / len(field_new))
    return field_aver, err
```
