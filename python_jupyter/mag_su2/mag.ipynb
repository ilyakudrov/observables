```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
df_old = pd.read_csv('/home/ilya/soft/lattice/general_code/apps/mag_su2/test/result/functional_thermalization_old.csv')
df_old['type'] = 'old'
df_spins = pd.read_csv('/home/ilya/soft/lattice/general_code/apps/mag_su2/test/result/functional_thermalization_test_spins.csv')
df_spins['type'] = 'spins'
df_vectors = pd.read_csv('/home/ilya/soft/lattice/general_code/apps/mag_su2/test/result/functional_thermalization_test_vectors.csv')
df_vectors['type'] = 'vectors'
df_vectors_fast = pd.read_csv('/home/ilya/soft/lattice/general_code/apps/mag_su2/test/result/functional_thermalization_test_vectors_fast.csv')
df_vectors_fast['type'] = 'vectors_fast'
df_test1 = pd.read_csv('/home/ilya/soft/lattice/general_code/apps/mag_su2/test/result/functional_thermalization_test.csv')
df_test1['type'] = 'test1'
print()
# df_spins = pd.read_csv('/home/ilya/soft/lattice/general_code/apps/mag_su2/test/result/functional_thermalization_vectors_fast.csv')
# df_spins['type'] = 'vectors_fast'
df = pd.concat([df_old, df_spins, df_vectors, df_vectors_fast, df_test1])
fg = seaborn.FacetGrid(data=df, hue='type', height=5,
                           aspect=1.6, legend_out=True)
fg.map(plt.errorbar, 'temperature', 'functional', mfc=None, fmt='o', ms=3, capsize=5, lw=0.5, ls=None
           ).add_legend()
fg.ax.set_xlabel('T', fontsize=14)
fg.ax.set_ylabel('functional', fontsize=14)
fg.ax.spines['right'].set_visible(True)
fg.ax.spines['top'].set_visible(True)
fg.ax.minorticks_on()
fg.ax.tick_params(which='both', bottom=True,
                  top=True, left=True, right=True)
```
