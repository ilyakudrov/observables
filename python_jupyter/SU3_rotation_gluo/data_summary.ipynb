```python
import pandas as pd
```

```python
df = pd.DataFrame({'A': [1, 1, 1, 2, 2], 'B': ['a', 'b', 'c', 'a', 'b'], 'C': [1, 2, 3, 4, 5]})
print(df)
df = df.set_index(['A', 'B']).unstack('B', fill_value=0).stack()
print(df)
```

```python
df = pd.read_csv('../../result/eos_rotation_imaginary/data_summary.csv')
print(df)
```

```python
# duplicated
df1 = df[df.duplicated(['lattice_size', 'boundary', 'velocity', 'beta'], keep=False)]
print(df1)
```

```python
def absent_parameters(df):
    df = df.set_index(['lattice_dir', 'lattice_size', 'boundary', 'velocity', 'beta']).unstack('beta', fill_value=0).stack()['observation_number'].reset_index()
    return df
```

```python
df1 = df[df['observation_number'] > 100]
df1 = df1[df1.duplicated(['lattice_size', 'boundary', 'velocity', 'beta'])]
df1 = df1.groupby(['lattice_dir', 'lattice_size', 'boundary']).apply(absent_parameters).reset_index(drop=True)
df1 = df1[df1['observation_number'] == 0]
print(df1)
df1.to_csv(f'../../result/eos_rotation_imaginary/beta_absent.csv', index=False)
```

```python
df1 = df[df['observation_number'] > 100]
df1 = df1[(df1['spec_bin_Pl.log'] != 'exists,not empty,readable') | (df1['spec_bin_S.log'] != 'exists,not empty,readable') | (df1['spec_therm.log'] != 'exists,not empty,readable')]
print(df1)
df1.to_csv(f'../../result/eos_rotation_imaginary/spec_absent.csv', index=False)
```
