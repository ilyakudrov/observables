```python
import numpy as np
import pandas as pd
import struct
import math
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import seaborn as sns
sns.set_style('darkgrid')
import sys
import os
import seaborn
from  matplotlib import pyplot
```

```python
def plot_potential_polyakov(path):
    df_data = pd.read_csv(f"../result/{path}/polyakov_correlator.csv", index_col=None)

    fg = seaborn.FacetGrid(data=df_data,  aspect=1.61)
    fg.map(plt.errorbar, 'distance', 'potential', 'err', marker="o", fmt='', linestyle = '').add_legend()

    image_path = f'../images/potential_polyakov/{path}'
    try:
        os.makedirs(image_path)
    except:
        pass

    image_path = f'{image_path}/potential_polyakov'
    fg.savefig(image_path)
```
