```python
import numpy as np
import pandas as pd
import struct
import math
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import seaborn as sns
sns.set_style('darkgrid')
import sys
import os
import seaborn
from  matplotlib import pyplot
```

```python
def plot_potential_polyakov(mu, monopole, lattice_size, smearing):
    df_data = pd.read_csv(f"../result/potential_polyakov/{monopole}/qc2dstag/{lattice_size}/potential_polyakov{smearing}_mu=0.{mu}.csv", index_col=None)
    
    fg = seaborn.FacetGrid(data=df_data,  aspect=1.61)
    fg.map(plt.errorbar, 'r/a', 'aV(r)', 'err', marker="o", fmt='', linestyle = '').add_legend()
    
    image_path = f'../images/potential_polyakov/{monopole}/qc2dstag/{lattice_size}'
    try:
        os.makedirs(image_path)
    except:
        pass
    
    image_path = f'{image_path}/potential_polyakov_mu=0p{mu}_time_dependence'
    fg.savefig(image_path)
```

```python
plot_potential_polyakov('00', '/', '32^4', '')
plot_potential_polyakov('00', 'monopoless', '32^4', '')
plot_potential_polyakov('00', 'monopole', '32^4', '')
```
