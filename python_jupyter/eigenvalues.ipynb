```python
def read_data_eigenvalues(mu):
    data = []
    for i in range(0, 1200):
        file_path = f"../data/eigenvalues/qc2dstag/24^4/mu{mu}/eigenvalues_{i:04}"

#         print(file_path)
        if(os.path.isfile(file_path)):
            data.append(pd.read_csv(file_path))
            data[-1]["conf_num"] = i

    df = pd.concat(data)
#     df = df[df['eigenvalue_real'] >= 0]
    
    return df
```

```python
df = read_data_eigenvalues('0.05')
# print(df)
df.plot.scatter(x='eigenvalue_real', y='eigenvalue_imaginary', s=0.001)
```
